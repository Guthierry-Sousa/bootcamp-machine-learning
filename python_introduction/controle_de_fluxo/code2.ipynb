### **Laços**

```python
pixels = [30, 120, 100, 45, 180, 200]
```

```python
# for + enumerate

for idx, p in enumerate(pixels):
    print(f"pixel[{idx}] = {p}")
```

```python
# while

idx = 0

while idx < len(pixels) and pixels[idx] < 150:

    idx+=1

print(f"Primeiro pixel ≥ 150 está na posição {idx}")
```

```python
# for

for p in pixels:

    if p == 45:
        continue
    if p > 190:
        break

    print(p)
```

```python
# List Comprehensions

pixels = [10, 20, 60, 40, 100, 190, 200, 255]

pixels_normalizados = [p/255.0 for p in pixels]

print(pixels_normalizados)
```

```python
saturados = [p for p in pixels if p > 150]

print(saturados)
```

```python
binarizado = [1 if p > 127 else 0 for p in pixels]

print(binarizado)
```
