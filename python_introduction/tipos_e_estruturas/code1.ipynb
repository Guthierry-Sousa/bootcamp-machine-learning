### **Tipos Primitivos**

Python é dinamicamente tipado: o tipo é inferido em tempo de execução.

```python
# Tipos Primitivos

inteiro = 42 # Int
flutuante = 3.14159 # float
booleano = True # bool
string = "Hello World" # str
```

```python
# Verificação e conversão

print(type(flutuante))
print(isinstance(inteiro, int))
print(float(inteiro))
```

```python
# Operações Básicas

print(255 // 2) # Divisão Inteira
print(255 / 2) #Divisão Normal
print(255 % 2) # Módulo
print(2 ** 8) # Potenciação
```
