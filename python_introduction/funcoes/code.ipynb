### **Funções em Python**

```python
def normalizar(pixel: int, maximo: int = 255) -> float:
    """Normaliza um valor de pixel para o intervalo [0, 1]."""

    if not  (0 <= pixel <= maximo):

        raise ValueError(f"Pixel esta fora do intervalo [0, {maximo}]")

    return pixel / maximo

def media(*valores: float) -> float:
    """Calcula a média dos valores"""

    return sum(valores) / len(valores)

def criar_config(**dados) -> dict:
    """Cria uma configuração inicial para uma imagem."""

    defaults = {'largura': 460, 'altura': 460, 'canais': 3}
    defaults.update(dados)

    return defaults
```

```python
print(normalizar(128))

print(media(0.2, 0.5, 0.8))

dados = {
    'largura': 690,
    'nome': 'hello'
}

print(criar_config(**dados))
```

### **Funções Lambda e Higher-Order Functions**

```python
# Lambda: função anônima de uma expressão

to_uint8 = lambda x: int(x * 255)

pixels_float = [0.0 , 0.5, 1.0]

# map - aplica função a cada elemento
pixels_uint8 = list(map(to_uint8, pixels_float))
print(pixels_uint8)
```

```python
pixels = [10, 130, 200, 80, 255]

# filter - filtra elementos por condição
claros = list(filter(lambda p: p > 127, pixels))
print(claros)
```

```python
imagens = [("img_c.png", 640), ("img_a.png", 1920), ("img_b.png", 320)]

ordenadas = sorted(imagens, key= lambda x: x[1])

print(ordenadas)
```
