#### **4.** Dada uma lista de números inteiros, escreva uma função para encontrar o segundo maior valor na lista.

```python
def segundo_maior_valor(lista: list[int]) -> int:

    """A função recebe uma lista e retorna o segundo maior valor"""

    if len(lista) <= 1:

        return None

    
    maior = float('-inf')
    segundo_maior = float('-inf')

    for i in lista:

        if i > maior:
            segundo_maior = maior
            maior = i
        elif (i > segundo_maior) and (i != maior):
            segundo_maior = i
    
    if segundo_maior == float('-inf'):

        return None

    return segundo_maior
```

```python
lista = [rd.randint(1, 100) for _ in range(10)]
print(f"Lista Original: {lista}")
```

```python
print(f"Segundo maior valor = {segundo_maior_valor(lista=lista)}")
```

```python
lista = []
print(f"Lista Original: {lista}")
```

```python
print(f"Segundo maior valor = {segundo_maior_valor(lista=lista)}")
```
