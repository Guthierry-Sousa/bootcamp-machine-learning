#### **3.** Escreva uma função que receba duas listas e retorne outra lista com os elementos que estão presentes em apenas uma das listas.

```python
def filtrar_elementos_unicos(lista1: list[int], lista2: list[int]) -> tuple[list[int], list[int]]:

    """A função recebe duas listas de inteiros e retorna duas listas com os inteiros exclusivos da lista 1 e os inteiros exclusivos da lista 2"""

    set_1, set_2 = set(lista1), set(lista2)

    return list(set_1 - set_2), list(set_2 - set_1)
```

```python
lista1 = [rd.randint(1, 20) for _ in range(10)]
print(f"Lista Original (Lista1): {lista1}")
```

```python
lista2 = [rd.randint(1, 20) for _ in range(10)]
print(f"Lista Original (Lista2): {lista2}")
```

```python
lista_unicos1, lista_unicos2 = filtrar_elementos_unicos(lista1=lista1, lista2=lista2)
print(f"Lista filtrada (apenas com números únicos da lista 1): {lista_unicos1}")

print(f"Lista filtrada (apenas com números únicos da lista 2): {lista_unicos2}")
```
