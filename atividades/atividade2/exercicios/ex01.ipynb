#### **1.** Escreva uma função que receba uma lista de números e retorne outra lista com os números ímpares.

```python
def filtro_impares(lista: list[int]) -> list[int]:
    
    """A função recebe uma lista e retorna outra lista apenas com os números ímpares."""

    return [x for x in lista if x%2 != 0]
```

```python
lista = [rd.randint(1, 100) for _ in range(10)]
print(f"Lista Original: {lista}")
```

```python
lista_impares = filtro_impares(lista=lista)
print(f"Lista filtrada (apenas com números ímpares): {lista_impares}")
```
