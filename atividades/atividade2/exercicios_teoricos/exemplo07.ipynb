### **Exemplo - Exercício 7**

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'B': [5, 6], 'C': [7, 8]})
df3 = pd.DataFrame({'D': [9, 10]})
```

##### **Exemplo 1**

```python
# Axis = 0 -> Linhas

resultado = pd.concat([df1, df2], axis=0)

display(resultado)
```

##### **Exemplo 2**

```python
# Axis = 1 -> Colunas

resultado = pd.concat([df1, df3], axis=1)

display(resultado)
```

- No **Exemplo 1,** o **df1** e **df2** foram utilizados com o **axis = 0** e possuem colunas diferentes, logo os espaços foram preenchidos como **NaN**.

- No **Exemplo 2**, o **df1** e **df3** foram utilizados com **axis = 1** e possuem o mesmo número de linhas, logo não ocorre os valores **NaN**.
