### **Exemplo - Exercício 10**

```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/data_exemplo_com_nulos.csv").drop(columns='id')
display(df.head())
print(df.shape)
```

```python
mask_valores_nan = df.isna()
df_nan = df[mask_valores_nan.any(axis=1)]

display(df_nan)
```

```python
qtd_values_nan = df.isna().sum()

print(f"Quantidade valores nulos:\n{qtd_values_nan}")
```

#### **Exclusão de valores nulos**

A remoção dos valores nulos pode ser uma opção de tratamento quando:

- A proporção dos valores nulos é pequena, que sua remoção não impactaria de forma significativa no resultado final.
- A origem dos dados nulos é do tipo MCAR (Missing Completely at Random), que sua remoção não iria introduzir viés no modelo.

```python
df_removed_nan = df.dropna()
display(df_removed_nan.head())
print(df_removed_nan.shape)
```

#### **Imputação de valores**

A imputação de valores pode ser uma opção de tratamento quando:
- A proporção dos valores nulos é muito grande.
- A origem dos dados nulos é do tipo MAR (Missing at Random).

Obs: A imputação também pode ser feita por meio de algoritmos de Machine Learning.

```python
df_imputed = df[['ram_gb', 'armazenamento_gb', 'nucleos_cpu', 'vram_gpu_gb', 'clock_cpu_ghz', 'preco_rs']].fillna(df[['ram_gb', 'armazenamento_gb', 'nucleos_cpu', 'vram_gpu_gb', 'clock_cpu_ghz', 'preco_rs']].mean())

df_imputed['marca'] = df['marca'].fillna(df['marca'].mode()[0])

display(df_imputed)
```
