#### **2.** Escreva uma função que receba uma lista de números e retorne outra lista com os números primos presentes.

```python
def filtro_numeros_primos(lista: list) -> list:
    """A função recebe uma lista e retorna outra lista apenas com os números primos"""

    lista_primos = list(filter(is_primo, lista))

    return lista_primos


def is_primo(num: int) -> bool:
    """A função recebe um inteiro e retona se ele é primo (True) ou não (False)"""
    count_div = 0
    for i in range(num, 1, -1):
        if num%i == 0:
            count_div += 1
        if count_div > 2:
            return False
    return True
```

```python
lista = [rd.randint(1, 100) for _ in range(10)]
print(f"Lista Original: {lista}")
```

```python
lista_primos = filtro_numeros_primos(lista=lista)
print(f"Lista filtrada (apenas com números primos): {lista_primos}")
```
