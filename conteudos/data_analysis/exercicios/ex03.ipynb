#### **3.** Escreva uma função que receba duas listas e retorne outra lista com os elementos que estão presentes em apenas uma das listas.

```python
def busca_binaria(lista: list, item: int) -> int:

    lista.sort()

    baixo = 0
    alto = len(lista) - 1

    while(baixo <= alto):

        meio = (baixo + alto) // 2

        if(lista[meio] == item):

            return True

        elif(item < lista[meio]):

            alto = meio - 1

        else:

            baixo = meio + 1

    return False 

def filtrar_elementos_unicos(lista1: list, lista2: list, unicos_lista1 = True) -> list:

    lista_unicos = []

    if unicos_lista1:

        lista_base = "Lista 1"

        for i in lista1:

            if not busca_binaria(lista2, i):

                lista_unicos.append(i)

    else:

        lista_base = "Lista 2"

        for i in lista2:

            if not busca_binaria(lista1, i):

                lista_unicos.append(i)

    return lista_unicos, lista_base
```

```python
lista1 = [rd.randint(1, 20) for _ in range(10)]
print(f"Lista Original (Lista1): {lista1}")
```

```python
lista2 = [rd.randint(1, 20) for _ in range(10)]
print(f"Lista Original (Lista2): {lista2}")
```

```python
lista_unicos, lista_base = filtrar_elementos_unicos(lista1=lista1, lista2=lista2, unicos_lista1=False)
print(f"Lista filtrada (apenas com números únicos da {lista_base}): {lista_unicos}")
```
