#### **Outliers**

```python
def identificar_outliers(data: pd.DataFrame, column: str) -> pd.DataFrame:

    series = data[column]

    Q3 = series.quantile(0.75)
    Q1 = series.quantile(0.25)

    IQR = Q3 - Q1

    lim_sup = Q3 + 1.5 * IQR
    lim_inf = Q1 - 1.5 * IQR

    outliers = (series < lim_inf) | (series > lim_sup)

    return data[outliers], lim_inf, lim_sup
```

```python
outliers, li, ls = identificar_outliers(df_clean, 'creativity_score')

display(outliers)
```
