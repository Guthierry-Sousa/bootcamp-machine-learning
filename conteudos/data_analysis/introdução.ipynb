#### **Leitura dos Dados**

```python
# Verificação de dados faltantes:
df.isna().sum()
```

```python
# Verificação de dados duplicados:

qtd_duplicados = int(df.duplicated().sum())

print(f"Dados Duplicados (Geral): {qtd_duplicados}")
```

```python
qtd_duplicados_id = int(df.duplicated(subset='projeto_id').sum())


print(f"Dados Duplicados ('projeto_id'): {qtd_duplicados_id}")
```

```python
# Verificação do número de valores únicos para cada coluna:

df.nunique()
```

```python
linguagens = list(df['linguagem'].unique())
print(f"Linguagens: {linguagens}")
```

```python
sistemas = list(df['sistema_operacional'].unique())
print(f"Sistemas Operacionais: {sistemas}")
```

```python
# Quais projetos usaram sistema operacional = 'Linux'?

display(df[df['sistema_operacional'] == 'Linux'])
```

```python
# Ordenação com base em uma coluna específica:

df.sort_values(by='commits', ascending=True).head(8)
```

```python
df.sort_values(by='commits', ascending=False).head(8)
```

```python
df[['desenvolvedores', 'area', 'linguagem', 'commits']].sort_values(by='commits', ascending=False).head(10)
```

#### **Plotagem de Gráficos**

##### **Variáveis Qualitativas**

```python
df['area'].value_counts()
```

```python
df['area'].value_counts().plot(kind='barh', color = 'orange')
plt.xlabel("Quantidade")
plt.grid(alpha = 0.35)
plt.show()
```

```python
df['linguagem'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

##### **Variáveis Quantitativas**

```python
df['desenvolvedores'].plot(kind='hist', color = 'green', bins = 5, edgecolor='black')
plt.ylabel("Desenvolvedores")
plt.show()
```

```python
df[df['desenvolvedores'] < 5].head(10)
```

```python
df['desenvolvedores'].plot(kind='kde', color = 'blue')
plt.show()
```

```python
sns.boxplot(df['desenvolvedores'])

plt.show()
```

#### **Análise de Correlação**

```python
plt.figure(figsize=(20, 15))
sns.heatmap(df.corr(numeric_only=True), cmap='BrBG', annot=True)
plt.show()
```
