### **Tipos Compostos**

```python
# Listas - Mutável

lista = [255, 128, 0, 64]
lista.append(32)
lista[0] = 200

print(lista)
```

```python
# Tupla - Imutável

shape = (480, 640, 3) # altura x largura x canais
altura, largura, canais = shape

print(f"altura = {altura}")
print(f"largura = {largura}")
print(f"canais = {canais}")
```

```python
# Dicionários - Mapeamento chave (key) - valor (value) 

metadados = {
    "formato": "PNG",
    "bits_por_canal": 8,
    "resolução": (72, 72)
}

print(metadados["formato"])
print(metadados["bits_por_canal"])
print(metadados["resolução"])

metadados["colorspace"] = "RGB"

print(metadados)
```
