## **Introdução - OpenCV**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### **Leitura e exebição de imagens**

```python
img = cv2.imread('img_exemplo.jpg') # por padrão, lê em BGR
```

```python
print(img)
```

```python
print(img.shape) # (altura, largura, canais)
print(img.dtype)
```

```python
plt.imshow(img)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converte para RGB
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
print(img_rgb)
```

### **Conversão**

```python
# Tons de Cinza:

img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
print(img_gray.shape)
```

```python
print(img_gray)
```

```python
# Preto e Branco:

limiar ,img_binary = cv2.threshold(img_gray, 127, 1, cv2.THRESH_BINARY)
plt.imshow(img_binary, cmap='gray')
plt.show()
```

```python
print(limiar)
```

```python
print(img_binary)
```

### **Acessando e manipulando pixels**

```python
pixel_exemplo = img_rgb[100, 50]
print(pixel_exemplo)
```

```python
img_rgb[100, 50] = [255, 255, 255]
print(img_rgb[100, 50])
```

### **Redimensionar Imagens**

```python
img_rgb_resized1 = cv2.resize(img_rgb, (100, 100)) # Quanto menor a resolução, menos detalhes a imagem vai possuir
plt.imshow(img_rgb_resized1)
plt.show()
```

```python
img_rgb_resized2 = cv2.resize(img_rgb, (1000, 800)) # Quanto maior a resolução, mais detalhes a imagem vai possuir
plt.imshow(img_rgb_resized2)
plt.show()
```

```python
print(f"shape resized1 = {img_rgb_resized1.shape}")
print(f"shape resized2 = {img_rgb_resized2.shape}")
```
