## **Tratamento de Valores Ausentes**

#### **Definição**: Valores ausentes (NaN) representam a ausência de um valor em uma determinada observação ou variável de um conjunto de dados.

### **Tipos de Dados Ausentes**

#### **MCAR - Missing Completely at Random**

- A probabilidade de um valor estar ausente é completamente independente de qualquer outra variável no dataset e do próprio valor que estaria presente.

**Exemplo**: Um pesquisador deixa cair acidentalmente alguns tubos de ensaio, perdendo as medições daquelas amostras.

**Possível Solução**: Remover valores MCAR é seguro, pois não introduz viés na análise. Mas ainda é ideal analisar a proporção desses valores no dataset.

#### **MAR - Missing at Random**

- A probabilidade de um valor estar ausente não depende do valor ausente em si, mas pode ser explicada por outras variáveis observadas no conjunto de dados.

**Exemplo**: Em uma pesquisa clínica, homens podem ser menos propensos a preencher um questionário sobre saúde mental do que mulheres. A ausência desse valores está relacionada à variável "gênero".

**Possível Solução**: Métodos de imputação que utilizam a informação das outras variáveis para prever os valores ausentes são mais apropriados e eficazes.

#### **MNAR - Missing Not at Random**

- A probabilidade de um valor estar ausente está diretamente relacionada ao valor que estaria presente. A ausência é sistemática e informativa, dependendo de fatores não observados.

**Exemplos**: Pessoas com rendas muito altas podem se sentir desconfortáveis em revelar seus ganhos e, portanto, são mais propensas a deixar a pergunta sobre renda em branco.

**Possível Solução**: Tanto a exclusão quanto os métodos de imputação padrão produzirão estimativas severamente enviesadas. O ideal é requerer a técnicas mais avançadas.

### **Estratégias de Tratamento**

- Exclusão de linhas e colunas;
- Imputação Simples;
- Imputação com Machine Learning;
- Utilizar algoritmos robustos a valores ausentes.
